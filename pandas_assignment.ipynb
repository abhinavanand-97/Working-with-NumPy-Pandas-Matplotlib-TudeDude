# **1. Create a DataFrame with the following data:**  
  
  data = {
      'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
      'Age': [24, 27, 22, 32, 29],
      'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
      'Salary': [45000, 54000, 50000, 62000, 47000]
  }

```python
# Importing Pandas
import pandas as pd

data = {
      'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
      'Age': [24, 27, 22, 32, 29],
      'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
      'Salary': [45000, 54000, 50000, 62000, 47000]
  }

# Converting Data into a Data-Frame using Pandas.
df = pd.DataFrame(data)
df
```

## a. Print the first five rows of the DataFrame.  

```python
'''
Since the dataset has exactly five rows, df.head() will display the entire DataFrame.
If we had more rows, it would show only the first five by default.
'''
# Print the first five rows of the DataFrame.
df.head(5)
```

##  b. Get the summary statistics of the 'Age' and 'Salary' columns.

```python
(df[['Age', 'Salary']].describe().round(2))

#This will display count, mean, std (standard deviation), min, max, and quartiles (25%, 50%, 75%) for both columns.
```

## c. Calculate the average salary of employees in the 'HR' department.

```python
avg_salary_hr = df[df['Department'] == 'HR']['Salary'].mean()
# This will filter only Alice and Eve (HR employees), then compute their average salary.

print("Average Salary of HR Department:", round(avg_salary_hr, 2))
```

# **2. Add a new column, 'Bonus', which is 10% of the salary.**

```python
df['Bonus'] = df['Salary']*0.1
print(df)
```

# **3. Filter the DataFrame to show employees aged between 25 and 30.**

```python
filtered_df = df[(df['Age'] >= 25) & (df['Age'] <= 30)]
filtered_df
```

# **4. Group the data by 'Department' and calculate the average salary for each department**

```python
dept_sal_avg = df.groupby('Department')['Salary'].mean()
dept_sal_avg
```

# **5. Sort the DataFrame by 'Salary' in ascending order and save the result to a new CSV file.**

```python
# Sort by Salary in ascending order
sorted_df = df.sort_values(by='Salary', ascending=True)

# Save to new CSV file
sorted_df.to_csv('sorted_salary.csv', index=False)

print(sorted_df)
```
